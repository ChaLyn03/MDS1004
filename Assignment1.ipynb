First, we will start by importing all relevant dependencies:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Then we will define each of the functions:

```python
def g(x):
    return (x+1)**2
def h(x):
    return (np.sqrt(x))-1
def p(x):
    return g(x)*h(x)
def r(x):
    return g(h(x))
def s(x):
    return h(g(x))
```

Now we can start plotting, here we will use numpy to create a vector of linear space between -3 and 3 with an interval of 0.1
We can now use this vector as the source of our x-values which is then fed into the function to calculate g(x)
Having now found x and g(x) for each value in the vector, we can plot x against g(x) to produce the below graph:

```python
x = np.arange(-3,3,0.1)
plt.plot(x,g(x))
```
